```python
def randomContractionMinCut(graphToEdit,seed=1):
    # select random node and one of its edges
    nodes = list(graphToEdit.keys())
    allEdges = [graphToEdit[nodeii] for nodeii in nodes]
    
    if len(nodes) <= 2: 
#         print(nodes)
#         print(graphToEdit[nodes[0]])
#         print(len(graphToEdit[nodes[0]]))
        return len(graphToEdit[nodes[0]])
    else:
        np.random.seed(seed)
        node = nodes[np.random.randint(0,len(nodes))]
        edges = graphToEdit[node]
        np.random.seed(seed)
        edge = edges[np.random.randint(0,len(edges))]
#         print(node,edge)
        # Find all edges to merge and remove edge
        edgesToAppend = graphToEdit[edge]
        edges = edges[edges!=edge]
        edges = np.append(edges,edgesToAppend)
        
        # Update the removed node's edges to point to new node
        for nodeToUpdate in edgesToAppend:
            if edge in graphToEdit[nodeToUpdate]:
                edgesToUpdate = graphToEdit[nodeToUpdate]
                edgesToUpdate[edgesToUpdate==edge] = node
                graphToEdit.update({nodeToUpdate:edgesToUpdate})

        # Remove any self loops
        edges = edges[edges!=node]
        
        # remove node corresponding to the edge selected
        graphToEdit.pop(edge) # pop the node, but the edge number is the node number
        
        
        # update graph
        graphToEdit.update({node:edges})
        minCut = randomContractionMinCut(graphToEdit,seed=seed)
        return minCut
```

```python
# Run it 100 times and select the min, should run n^2  which is 4000 = 200^2
minCutVals = []
for seed in range(100):
    # My graph answer
    graph = {}
    with open('kargerMinCut.txt') as f:
        lines = f.readlines()
        for line in lines:
            nodeEdges = list(map(int,line.split('\t')[:-1]))
            graph.update({nodeEdges[0]:np.asarray(nodeEdges[1:])})

    graphToEditNew = dict(graph)
    numberOfEdges = randomContractionMinCut(graphToEditNew,seed=seed)
    print(numberOfEdges)
    minCutVals.append(numberOfEdges)
min(minCutVals)
```
